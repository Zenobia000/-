```python
import json
import pandas as pd
```

```python
with open('job7_v2.json', 'r', encoding='utf-8') as infile:
    jobs7 = json.load(infile)
```

```python
salary = []
tool = []

for i in range(len(jobs7)):
    salary.append(jobs7[i]["工作待遇"])
    tool.append(jobs7[i]["擅長工具"])
```

```python
tool_salary = pd.DataFrame({
    'salary': salary,
    'tool': tool
})
```

```python
tool_salary
```

```python
tool_salary_drop = tool_salary.loc[tool_salary['salary'] != '待遇面議（經常性薪資達 4 萬元或以上）']
```

```python
tool_salary_drop
```

```python
tool_salary_drop2 = tool_salary_drop.assign(tool=tool_salary_drop.tool.array).explode('tool')
```

```python
tool_salary_drop2
```

```python
tool_salary_drop2.to_csv('tool_salary_drop2.csv', encoding='utf-8')
```
