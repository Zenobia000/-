```python
import json
import pandas as pd
```

```python
with open('job7_v2.json', 'r', encoding='utf-8') as infile:
    jobs7 = json.load(infile)
```

```python
city = []
city_str = []
district = []
district_str = []
salary = []

for i in range(len(jobs7)):
    city.append(jobs7[i]["縣市編碼"])
    city_str.append(jobs7[i]["縣市"])
    district.append(jobs7[i]["地區編碼"])
    district_str.append(jobs7[i]["地區"])
    salary.append(jobs7[i]["工作待遇"])
```

```python
location_salary = pd.DataFrame({
    'city': city,
    'city_str': city_str,
    'district':district,
    'district_str':district_str,
    'salary': salary
})
```

```python
location_salary
```

```python
location_salary_drop = location_salary.loc[location_salary['salary'] != '待遇面議（經常性薪資達 4 萬元或以上）']
```
