```python
import json
import pandas as pd
```

```python
with open('job7_v1.json', 'r', encoding='utf-8') as infile:
    jobs7 = json.load(infile)
```

```python
# 薪水轉整數 濾掉論件計酬
```

```python
for i in range(len(jobs7)-1, -1, -1):
    
    salary = jobs7[i]["工作待遇"]
    if isinstance(salary, float):
        jobs7[i]["工作待遇"] = int(salary)
    elif isinstance(salary, str) and salary[:4] == '論件計酬':
        del jobs7[i]
```

```python
#整理學歷
```

```python
# '高中以下', '高中', '專科', '大學', '碩士', '博士'

idn = []
salary = []
education = []

for i in range(len(jobs7)):
    edu = jobs7[i]["學歷要求"]
    edu_list = edu.split("、")
    edu_min = edu_list[0]
    
    if edu_min[-2:] == '以上':
        edu_min = edu_min[:-2]
    
    edu = jobs7[i]["學歷要求"] = edu_min
    
    idn.append(jobs7[i]["序號"])
    salary.append(jobs7[i]["工作待遇"])
    education.append(jobs7[i]["學歷要求"])
```

```python
edu_salary = pd.DataFrame({
    '序號':idn, 
    '薪水':salary, 
    '學歷要求':education})
```

```python
edu_salary
```

```python
edu_salary_drop = edu_salary.loc[edu_salary['薪水'] != '待遇面議（經常性薪資達 4 萬元或以上）']
```

```python
edu_salary_drop 
```

```python
edu_salary_drop = edu_salary_drop.dropna()
```

```python
edu_salary_drop.to_csv('edu_salary_drop.csv', encoding='utf-8')
```
