```python
#整理擅長工具
```

```python
r = r'、|贊助提升專業能力'
r_s_1 = r'\s'
r_s = r'\s+'

for i in range(len(jobs7)):
    tool = jobs7[i]["擅長工具"]
    if tool == '不拘':
        jobs7[i]["擅長工具"] = None
        continue
    elif tool == None:
        continue
    tool = re.sub(r_s_1, "", tool)
    tool = re.sub(r," ", tool)
    tool = re.sub(r_s, " ", tool)
    tool = ''.join(alp for alp in tool if alp.isprintable())
    tool = tool.split()
    jobs7[i]["擅長工具"] = tool
```

```python
idn = []
tools = []
for i in range(len(jobs7)):
    idn.append(i)
    tools.append(jobs7[i]['擅長工具'])
    
    #加index進去json檔裡
    jobs7[i]['序號']=i
```

```python
tool_df3 = tool_df2['tool'].str.join('|').str.get_dummies()
```

```python
group_std = []
```

```python
for i in range(2, 15):
    KM = KMeans(n_clusters=i, init='random', random_state=5)
    KM.fit(tool_df3)
    c = KM.cluster_centers_
    group_std.append(np.std(c))
    print(i, " ", np.std(c))
```

```python
kmeans_std = pd.DataFrame({
    'std': group_std
    },
    index = [i for i in range(2, 15)]
)
```

```python
def convert_num(str_d):
    num_d = 0
    for s in str_d:
        if s != ',':
            num_d = num_d*10 + int(s)
    return num_d
```

```python
r_d = r'\d{1,3}[,\d{3}]*'
for i in range(len(jobs7)):
    salary = jobs7[i]["工作待遇"]
    if salary != None and salary != '待遇面議（經常性薪資達 4 萬元或以上）':
                    
        s_str = re.findall(r_d, salary) #抓數字字串
        s_avg = 0
                    
        if s_str == []: #沒數字
            continue
        elif len(s_str) == 1: #一個數字
            s_avg = convert_num(s_str[0])
        elif len(s_str) == 2: #兩個數字
            s_avg = (convert_num(s_str[0]) + convert_num(s_str[1])) / 2
        else: #error
            continue
                        
        if salary[:2] == '年薪':
            jobs7[i]["工作待遇"] = s_avg/13
        elif salary[:2] == '月薪':
            jobs7[i]["工作待遇"] = s_avg
        elif salary[:2] == '日薪':
            jobs7[i]["工作待遇"] = s_avg*22
        elif salary[:2] == '時薪':
            jobs7[i]["工作待遇"] = s_avg*174
        else: #error
            continue
```
